```python
#necessary header files required
import numpy as np # linear algebra
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("./Train_Test_data/train.csv")
df_test = pd.read_csv("./Train_Test_data/test.csv")
df_train.head()
```

```python
def funtion(x):
    x=x.strip()
    if(x=="Dimorphic hemmorhoids(piles)"):
        x="Dimorphic hemorrhoids(piles)"
    return x
df_train["Disease"]=df_train["Disease"].apply(funtion)
df_test["Disease"]=df_test["Disease"].apply(funtion)
```

```python
#removing the unnecessary col from data
df_train = df_train.iloc[:,1:]
df_test = df_test.iloc[:,1:]
df_train.head()
df_test.head()
```

```python
#splitting the data of train and test set for prediction
X_train = df_train.iloc[:, :-1]
X_test =  df_test.iloc[:, :-1]
y_train = df_train.iloc[:, -1:]
y_test =  df_test.iloc[:, -1:]
```

```python
print("X train shape: ",X_train.shape)
print("y train shape: ",y_train.shape)
```

```python
# Building the ANN MOdel
#firstly, we will just build a simple NN model, with two layers capable of solving linear classification problem.

class NeuralNet(object):
    def __init__(self):
        # Generate random numbers
        np.random.seed(1)
        
        # Assign random weights to a 3 * 1 matrix
        self.synaptic_weights = 2 * np.random.random((3, 1)) - 1
        
    # The sigmoid function method
    def _sigmoid(self,x):
        return 1 / (1 + np.exp(-x))
    
    # Derivative sigmoid
    def derivative_sigmoid(self, x):
        return x * (1 - x)
    
    # Train the neural network and adjust the weights each time
    def train(self, inputs, outputs, iteration_number):
        for iteration in range(iteration_number):     
        
            # Pass the training set through network
            output = self.learn(inputs)
        
            # Calculate the error
            error = outputs - output
        
            # Adjust the weights by a factor
            factor = np.dot(inputs.T, error * self.derivative_sigmoid(output))
            self.synaptic_weights += factor
        
    # calculate z    
    def learn(self, test_inputs):
        return self._sigmoid(np.dot(test_inputs, self.synaptic_weights))
```

```python
# Initialize and use NN
neural_net = NeuralNet()

# The training set
inputs = np.array([[0, 1, 1], [1, 0, 0], [1, 0, 1]])
outputs = np.array([[1, 0, 1]]).T

# train the neural network
neural_net.train(inputs, outputs, 50)

test_inputs = np.array([1, 0, 1])
threshold = 0.5
if neural_net.learn(test_inputs) >= threshold:
    print("Our test example output is: 1 hurray!! executed successfully!!!")
else:
    print("Our test example output is: 0")
```

```python
y_train_dum = pd.get_dummies(y_train)
y_train_dum
```

```python
# import tensorflow and keras
import tensorflow as tf
from tensorflow.keras.models import Sequential   # used for initialize ANN model
from tensorflow.keras import layers   # used for different layer structure
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping 
```

```python
#initialsing the ANN Model
model = Sequential()
# adding first hidden layer with input layer. there is parameter that represents how to initialize weights
model.add(Dense(64, activation = "relu", input_dim = X_train.shape[1]))

# adding second hidden layer
model.add(Dense(32, activation = "relu"))

#addding third hidden layer
model.add(Dense(16,activation="relu"))

# adding last layer
model.add(Dense(y_train_dum.shape[1], activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile('adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train_dum, epochs = 5, batch_size = 30)
```

```python
#Prediction
prediction = model.predict_classes(X_test)
#prediction=np.argmax(model.predict(X_test), axis=-1)
prediction
```
