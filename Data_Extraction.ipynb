```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter("ignore")
```

```python
data=pd.read_csv("./raw_data/dataset.csv")
#imported the file and saved in the data dataframe
```

```python
data.head()
# visiting the data set of the files 
```

```python
data.describe()
```

```python
data["Symptoms"]=0
records = data.shape[0]
for i in range(records):
    values = data.iloc[i].values
    values = values.tolist()
    if 0 in values:
        data["Symptoms"][i] = values[1:values.index(0)]
    else:
        data["Symptoms"][i] = values[1:]
data.head()
```

```python
#we are needed to craete the data frame of the column
column_values = data[['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4',
       'Symptom_5', 'Symptom_6', 'Symptom_7', 'Symptom_8', 'Symptom_9',
       'Symptom_10', 'Symptom_11', 'Symptom_12', 'Symptom_13', 'Symptom_14',
       'Symptom_15', 'Symptom_16', 'Symptom_17']].values.ravel()
symps = pd.unique(column_values)
symps = symps.tolist()
symps = [i for i in symps if str(i) != "nan"]
```

```python
symptoms = pd.DataFrame(columns = symps,index = data.index)
symptoms["Symptoms"] = data["Symptoms"]
for i in symps:
    symptoms[i] = symptoms.apply(lambda x:1 if i in x.Symptoms else 0, axis=1)
symptoms["Disease"] = data["Disease"]
symptoms = symptoms.drop("Symptoms",axis=1)
symptoms.head()
```

```python
def max_value(x):
    mx=-1e6
    for i in x:
        mx=max(mx,i)
    return mx
def min_value(x):
    mn=1e6
    for i in x:
        mn=min(mn,i)
    return mn
        
```

```python
max_sever=max_value(sever["weight"])
min_sever=min_value(sever["weight"])
print(max_sever,min_sever)
```

```python
def siverarity_ratio(x):
    return float(x/10)
sever["weight"]=sever["weight"].apply(siverarity_ratio)
sever["weight"]
```
